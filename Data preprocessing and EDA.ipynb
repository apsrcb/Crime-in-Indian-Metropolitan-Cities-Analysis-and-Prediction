# DATA PROCESSING

```python
# With pandas, one can efficiently analyze, manipulate, and process data 
import pandas as pd
```

```python
# pd.read_csv() function in pandas is used to read data from a CSV file and create a DataFrame object.

data_eo = pd.read_csv(r'C:\Users\USER\Desktop\data\Economic Offences.csv')
data_w = pd.read_csv(r'C:\Users\USER\Desktop\data\Crime Against Women .csv')
data_cc = pd.read_csv(r'C:\Users\USER\Desktop\data\Cyber Crime.csv')
data_v = pd.read_csv(r'C:\Users\USER\Desktop\data\Violent Crimes.csv')
data_ka = pd.read_csv(r'C:\Users\USER\Desktop\data\Kidnapping and Abduction.csv')
data_sll = pd.read_csv(r'C:\Users\USER\Desktop\data\SLL.csv')
data_ipc = pd.read_csv(r'C:\Users\USER\Desktop\data\IPC.csv')
```

#### Merging multiple datasets into a single dataset

```python
# pd.merge() function is used to combine data from multiple DataFrames based on a common key or keys.
# 'common_key' - 'Year' and 'City'

data = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(data_eo, data_w, on=['City', 'Year'], how='inner'), data_cc, on=['City', 'Year'], how='inner'), data_v, on=['City', 'Year'], how='inner'), data_ka, on=['City', 'Year'], how='inner'), data_sll, on=['City', 'Year'], how='inner'), data_ipc, on=['City', 'Year'], how='inner')
```

```python
# to_csv() function in pandas is used to save a DataFrame object as a CSV file

data.to_csv('dataset_data.csv', index=False)
```

### handling outliers

```python
import numpy as np

# Select multiple desired attributes
attributes = ['eo cases', 'w cases', 'cc cases', 'violent cases', 'sll cases', 'ipc cases']  

# Calculate the Z-score for each data point in the selected columns
z_scores = data[attributes].apply(lambda x: (x - x.mean()) / x.std())

# Set a threshold for outlier detection (e.g., Z-score > 3 or Z-score < -3)
threshold = 3

# Identify outliers
outliers = data[np.abs(z_scores) > threshold]

# Display the outliers
print(outliers)
```

#### no outliers found
